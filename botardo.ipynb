```python
import pandas as pd
from pandas import ExcelWriter

from google.colab import drive
drive.mount('/content/drive')
```

exporto datos a un excel

```python
"exporto una columna con productos seguido de varios precios del mismo, son 4 columnas"
productos = ["tomate cherry", "tomate platense", "tomate perita", "kiwi"]
precios_minimos = [2000,500,400,1500]
precios_modales = [2500,750,450,2000]
precios_maximos = [3000,1000,600,2250]

PATH = "/content/drive/MyDrive"
FILE_NAME = "Precio productos varios"
name_columns = ["producto","precio minimo","precio modal","precio maximo"]
df = pd.DataFrame(list(zip(productos,precios_minimos,precios_modales,precios_maximos)), columns=name_columns)
path = ExcelWriter(PATH + "/" + FILE_NAME + ".xlsx", mode="w")
df.to_excel(path, index=False, startrow = 0, startcol = 0) # con index en false, si esta en true numera cada fila
path.close()
```
