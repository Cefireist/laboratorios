```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Obtenemos los datos
datos1 = np.array([1.428, 1.426, 1.428, 1.428, 1.428, 1.424, 1.426, 1.426, 1.428, 1.428])
datos2 = np.array([1.526, 1.524, 1.523, 1.536, 1.526, 1.529, 1.525, 1.527, 1.528, 1.525])
```

```python
# Si queremos hacer un solo histograma, usamos esto
cant_datos_1 = len(datos1)
alturas, bins = np.histogram(datos1, bins="sqrt")
# divido la altura de cada barra por la cantidad de datos para obtener frecuencias
frecuencias = alturas/cant_datos_1

ancho_de_bin = bins[1] - bins[0]
plt.bar(bins[:-1], frecuencias, align='edge', width = ancho_de_bin, label='histograma', color='yellow', edgecolor='black', alpha = 0.5 )
plt.legend()
plt.grid(axis='y')
plt.xlabel('magnitud', fontsize=15)
plt.ylabel('densidad de probabilidad', fontsize=15)
# plt.xticks(bins) # esto es para poner los valores en los bordes de los bins si se quisiera fijarlos

# Mostramos los parametros estadisticos
```

```python
# Valor mas representativo (promedio):

p1 = np.mean(datos1)
# Desviaciones estandar:
s1 = np.std(datos1)
# Errores estadisticos oe = s/raiz(n):
oe1 = s1/cant_datos_1

print("Valor mas representativo p1: " + str(p1))
print("Desviacion estandar s1: " + str(s1))
print("sigma estadistico σe1: " + str(oe1))
```

```python
# Si queremos hacer 2 histogramas juntos podemos usar esto
cant_datos_1 = len(datos1)
cant_datos_2 = len(datos2)
alturas, bins = np.histogram(datos1, bins="sqrt")
alturas2, bins2 = np.histogram(datos2, bins='sturges')
# divido la altura de cada barra por la cantidad de datos para obtener frecuencias
frecuencias = alturas/cant_datos_1
frecuencias2 = alturas2 / cant_datos_2

ancho_de_bin = bins[1] - bins[0]
ancho_de_bin2 = bins2[1] - bins2[0]
# Graficamos
plt.bar(bins[:-1], frecuencias, align='edge', width=ancho_de_bin, alpha=0.9, label='histograma 1', color='purple', edgecolor='black' )
plt.bar(bins2[:-1], frecuencias2, align='edge', width=ancho_de_bin2, alpha=0.6, label='histograma 2', color='orange', edgecolor='black' )
# Parametros para el grafico, texto
plt.legend()
plt.grid(True)
plt.xlabel('Magnitud [ud]', fontsize=15)
plt.ylabel('densidad de probabilidad', fontsize=15)
# plt.xticks(bins) # esto es para poner los valores en los bordes de los bins si se quisiera fijarlos, tambien podria ir un array p. ej [1,2,3,4,5]
plt.show()
```

```python
# Vamos a obtener parametros estadisticos:

#Valores mas representativos (promedio):
p1 = np.mean(datos1)
p2 = np.mean(datos2)
# Desviaciones estandar:
s1 = np.std(datos1)
s2 = np.std(datos2)
# Errores estadisticos oe = s/raiz(n):
oe1 = s1/np.sqrt(cant_datos_1)
oe2 = s2/np.sqrt(cant_datos_2)
# Mostramos los valores:
print("Valor mas representativo p1: " + str(p1))
print("Desviacion estandar s1: " + str(s1))
print("sigma estadistico σe1: " + str(oe1))
print("")
print("Valor mas representativo p2: " + str(p2))
print("Desviacion estandar s2: " + str(s2))
print("sigma estadistico σe2: " + str(oe2))
```
